```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("Position_Salaries.csv")
X=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X,y) # type: ignore
```

```python
from sklearn.preprocessing import PolynomialFeatures
pr=PolynomialFeatures(degree=10)
X_poly=pr.fit_transform(X)
lr_2=LinearRegression()
lr_2.fit(X_poly,y) # type: ignore
```

```python
plt.scatter(X,y,color="red") # type: ignore
plt.plot(X,lr.predict(X),color="blue")
plt.title("Truth Linear Regression")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(X,y,color="red") # type: ignore
plt.plot(X,lr_2.predict(pr.fit_transform(X)),color="blue")
plt.title("Truth Polynomial Regression")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```
