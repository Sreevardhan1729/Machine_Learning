```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Mall_Customers.csv")
X=df.iloc[:,3:].values
```

```python
from sklearn.cluster import KMeans
wss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init="k-means++",random_state=42)
    kmeans.fit(X)
    wss.append(kmeans.inertia_) #we call intertia for finding best case for the elbow method
plt.plot(range(1,11),wss)
plt.title("elbow method")
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init="k-means++",random_state=42)
y_k=kmeans.fit_predict(X)
y_k
```

```python
plt.scatter(X[y_k==0,0],X[y_k==0,1],s=100,color="red",label="cluster1")
plt.scatter(X[y_k==1,0],X[y_k==1,1],s=100,color="cyan",label="cluster2")
plt.scatter(X[y_k==2,0],X[y_k==2,1],s=100,color="yellow",label="cluster3")
plt.scatter(X[y_k==3,0],X[y_k==3,1],s=100,color="blue",label="cluster4")
plt.scatter(X[y_k==4,0],X[y_k==4,1],s=100,color="magenta",label="cluster5")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,color="green",label="Centroids")
plt.legend()
plt.show()
```
