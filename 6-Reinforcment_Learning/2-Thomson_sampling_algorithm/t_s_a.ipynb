```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Ads_CTR_Optimisation.csv")
```

```python
import random
N=250
d=10
ads_selected=[]
rewards_1=[0]*d
rewards_0=[0]*d
total_rewards=0
for n in range(N):
    ad=0
    max_random=0
    for i in range(d):
        random_beta=random.betavariate(rewards_1[i]+1,rewards_0[i]+1)
        if(max_random<random_beta):
            max_random=random_beta
            ad=i
    ads_selected.append(ad)
    reward=df.values[n,ad]
    if reward==1:
        rewards_1[ad]+=1
    else:
        rewards_0[ad]+=1
    total_rewards+=reward
```

```python
plt.hist(ads_selected)
plt.show()
```
